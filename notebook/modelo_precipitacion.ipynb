#### Ajuste de modelos para la predicción de la precipitación

```python
import pandas as pd
import plotly.express as px
```

```python
# datya del parque
polygonName = 'Cerro Saroche'

data_parque = pd.read_pickle('./data/{polygonName}_data.pkl'.format(polygonName=polygonName))
data_parque.head(5)
```

```python
# viendo los datos
print(data_parque.date.min())
print(data_parque.date.max())
```

```python
fig = px.line(data_parque[data_parque.id==0],
                    y='total_precipitation',
                    x='date_time',
                    #color='total_precipitation',
                    title='Precipitación')
fig.show()
```
