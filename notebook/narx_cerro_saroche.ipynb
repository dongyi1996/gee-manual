```python
def narx_p1(output_order,input_order,hidden_nodes):

    # data
    all_inputs = Y[:-1]
    all_targets = Y[1:]

    # configuracion de la red
    input_nodes = 1
    hidden_nodes = hidden_nodes
    output_nodes = 1

    output_order = output_order
    incoming_weight_from_output = .5
    input_order = input_order
    incoming_weight_from_input = .5

    net = NeuralNet()
    net.init_layers(input_nodes, [hidden_nodes], output_nodes,
            NARXRecurrent(
                output_order,
                incoming_weight_from_output,
                input_order,
                incoming_weight_from_input))

    net.randomize_network()

    # fijando data
    net.set_all_inputs(all_inputs) 
    net.set_all_targets(all_targets)

    #
    length = len(all_inputs)
    years_predict = 1

    net.set_learn_range(0, int( ( length-(12*years_predict) ) ) -1 ) 
    net.set_test_range(int( ( length-(12*years_predict) ) )- 1, length -1  ) 

    # funcion activacion
    #net.layers[1].set_activation_type('sigmoid')
    net.layers[1].set_activation_type('tanh')

    net.learn(epochs=10, 
             show_epoch_results=False,
             random_testing=True)

    return net.test()
```

```python
output_order = list(range(21,30+1))
input_order = list(range(11,20+1))
hidden_nodes = list(range(21,30+1))

list_mse = list(map(lambda x,y,z: [x,y,z,narx_p1(x,y,z)],
                                    output_order,
                                    input_order,
                                    hidden_nodes
                                    ))

resultados_1 = pd.DataFrame(list_mse,columns=['output_order','input_order','hidden_nodes','mse'])
```
